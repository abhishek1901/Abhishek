```
import matplotlib.pyplot as plt
import sklearn.datasets as datasets
import pandas as pd
import numpy as np
```

```
iris=datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```
#To find the Value of K for K Means Clustering
#Importing KMeans
from sklearn.cluster import KMeans
#Normalizing the Data
x = df.iloc[:,:].values

sum_of_squared_distances = []

for i in range(1,10):
    k = KMeans(init = "random", n_clusters=i, n_init =10, max_iter=300, random_state=30)
    k.fit(x)
    sum_of_squared_distances.append(k.inertia_)


plt.title("Elbow Method to find the Value of Optimal K")
plt.xlabel("No. of Clusters")
plt.ylabel("Sum of Squared Distance")
plt.plot(range(1,10),sum_of_squared_distances)
```

```
#From the above Graph it is Evident that the Optimum Value for K here is 3
k = KMeans(init = "random", n_clusters=3, n_init =10, max_iter=300, random_state=30)
k_pred = k.fit_predict(x)
```

```
#Plotting the Result
plt.scatter(x[k_pred == 0, 0], x[k_pred == 0, 1], 
            s = 200, c = 'red', label = 'Iris-setosa')
plt.scatter(x[k_pred == 1, 0], x[k_pred == 1, 1], 
            s = 200, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[k_pred == 2, 0], x[k_pred == 2, 1],
            s = 200, c = 'green', label = 'Iris-virginica')
plt.scatter(k.cluster_centers_[:,0], k.cluster_centers_[:,1], 
            s = 200, c = 'yellow', label = 'Centroids')
plt.legend()
```
